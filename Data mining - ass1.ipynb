```python
# Normalize numerical features using Min-Max scaling or Z-score standardization with sklearn.preprocessing.
df_numeric = df.select_dtypes(include=['float64', 'int64'])

scaler = MinMaxScaler()
pd.DataFrame(scaler.fit_transform(df_numeric), columns=df_numeric.columns)
```

```python
# Use pd.cut() to bin continuous variables into discrete intervals.
bins=[0, 40, 60, 80, 100]
labels=['Low', 'Medium', 'High', 'Very High']
df['Temperature_Binned'] = pd.cut(df['Temperature'], bins=bins, labels=labels)
df
```

```python
# Exercise 4: Feature Engineering
# Create new features by combining or transforming existing features (e.g., adding interaction terms or polynomial features).

df['Temp_Fuel_Interaction'] = df['Temperature'] * df['Fuel_Price']
poly = PolynomialFeatures(degree=2, include_bias=False)
temp_poly = poly.fit_transform(df[['Temperature']])
df_temp_poly = pd.DataFrame(temp_poly, columns=['Temp', 'Temp^2'])
df = pd.concat([df, df_temp_poly], axis=1)
df
```

```python
# Extract date-based features (e.g., year, month, day) from datetime columns using pd.to_datetime() and dt accessor.
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name() 
df
```

```python
# Use domain knowledge to engineer features that might be useful for your specific problem.
df['Is_Summer'] = df['Date'].dt.month.isin([6, 7, 8]).astype(int)
df['Is_Winter'] = df['Date'].dt.month.isin([12, 1, 2]).astype(int)
df['Is_Spring'] = df['Date'].dt.month.isin([3, 4, 5]).astype(int)
df['Is_Fall'] = df['Date'].dt.month.isin([9, 10, 11]).astype(int)
# feature indicating optimal temperature range for sales
df['Optimal_Temp'] = ((df['Temperature'] >= 60) & (df['Temperature'] <= 75)).astype(int)
# feature for high fuel prices
df['High_Fuel_Price'] = (df['Fuel_Price'] > 4.0).astype(int)
df.head()
```
